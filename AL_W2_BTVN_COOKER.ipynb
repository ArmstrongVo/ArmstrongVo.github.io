```python
from matplotlib import figure
from skfuzzy.membership.generatemf import gaussmf
import numpy as np
import skfuzzy as fuzz
import matplotlib.pyplot as plt
from skfuzzy import control as ctrl 
time=ctrl.Antecedent(np.arange(10,180,0.1),'time')
mass=ctrl.Antecedent(np.arange(0.2,1.5,0.1),'mass')
power=ctrl.Consequent(np.arange(0,100,0.1),'power')
#tạo ra 1 tín hiệu input từ 10-180 với độ phân giải là 0.1
#tạo ra 1 tín hiệu output từ 0.2-1.5 với độ phân giải là 0.1
# tự động bằng câu lệnh automsf(n)_ n là số lẻ
time['VF']=fuzz.trimf(time.universe,[10,15,20])
time['F']=fuzz.trimf(time.universe,[15,20,35])
time['AVG']=fuzz.trimf(time.universe,[20,35,45])
time['S']=fuzz.trimf(time.universe,[35,45,120])
time['VS']=fuzz.trimf(time.universe,[45,120,180])
# VF - very fast
# F - fast
# AVG -average
# S - slow
# VS - very slow

mass['FE']=fuzz.trimf(mass.universe,[0.2,0.3,0.4])
mass['L']=fuzz.trimf(mass.universe,[0.3,0.4,0.5])
mass['AVG']=fuzz.trimf(mass.universe,[0.4,0.5,0.8])
mass['M']=fuzz.trimf(mass.universe,[0.5,0.8,1.1])
mass['VM']=fuzz.trimf(mass.universe,[0.8,1.1,1.5])
# FE - few
# L -little
# AVG - average
# M - much
# VM - very much

power['VSM']=fuzz.trimf(power.universe,[0,10,20])
power['SM']=fuzz.trimf(power.universe,[10,20,50])
power['AVG']=fuzz.trimf(power.universe,[20,50,70])
power['B']=fuzz.trimf(power.universe,[50,70,90])
power['VB']=fuzz.trimf(power.universe,[70,90,100])
# VSM -very small
# SM - small
# AVG - average
# B - big
# VB - very big  

time.view()      
mass.view()   
power.view()            
rule1=ctrl.Rule(time['VF']&mass['FE'],power['B'])  
rule2=ctrl.Rule(time['F']&mass['FE'],power['AVG'])    
rule3=ctrl.Rule(time['AVG']&mass['FE'],power['AVG'])  
rule4=ctrl.Rule(time['S']&mass['FE'],power['VSM']) 
rule5=ctrl.Rule(time['VS']&mass['FE'],power['VSM'])  

rule6=ctrl.Rule(time['VF']&mass['L'],power['B'])    
rule7=ctrl.Rule(time['F']&mass['L'],power['AVG'])  
rule8=ctrl.Rule(time['AVG']&mass['L'],power['AVG']) 
rule9=ctrl.Rule(time['S']&mass['L'],power['SM'])  
rule10=ctrl.Rule(time['VS']&mass['L'],power['VSM']) 

rule11=ctrl.Rule(time['VF']&mass['AVG'],power['B'])  
rule12=ctrl.Rule(time['F']&mass['AVG'],power['B']) 
rule13=ctrl.Rule(time['AVG']&mass['AVG'],power['AVG'])  
rule14=ctrl.Rule(time['S']&mass['AVG'],power['SM'])    
rule15=ctrl.Rule(time['VS']&mass['AVG'],power['VSM']) 

rule16=ctrl.Rule(time['VF']&mass['M'],power['VB']) 
rule17=ctrl.Rule(time['F']&mass['M'],power['VB']) 
rule18=ctrl.Rule(time['AVG']&mass['M'],power['B'])  
rule19=ctrl.Rule(time['S']&mass['M'],power['AVG'])    
rule20=ctrl.Rule(time['VS']&mass['M'],power['SM']) 

rule21=ctrl.Rule(time['VF']&mass['VM'],power['VB']) 
rule22=ctrl.Rule(time['F']&mass['VM'],power['VB'])  
rule23=ctrl.Rule(time['AVG']&mass['VM'],power['B'])    
rule24=ctrl.Rule(time['S']&mass['VM'],power['AVG'])  
rule25=ctrl.Rule(time['VS']&mass['VM'],power['SM'])

#rule1.view()
#rule2.view()
power_ctrl=ctrl.ControlSystem([rule1,rule2,rule3,rule4,rule5,rule6,rule7,rule8,rule9,rule10,rule11,rule12,rule13,rule14,rule15,rule16,rule17,rule18,rule19,rule20,rule21,rule22,rule23,rule24,rule25])
powercontrol = ctrl.ControlSystemSimulation(power_ctrl)
powercontrol.input['time']=35
powercontrol.input['mass']=1.1
powercontrol.compute()
print(powercontrol.output['power'])
power.view(sim=powercontrol)
  
```
