```python
from matplotlib import figure
#from skfuzzy.membership.generatemf import gaussmf
import matplotlib.pyplot as plt
#from skfuzzy import control as ctrl 
import numpy as np
#import skfuzzy as fuzz
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import axis3d
import pandas as pd
url = 'Salary_Data.csv'
data = pd.read_csv(url)
data.hist()
plt.show 
```

```python
#correlation = data.corr() # mối tương quan giữa các biến, trị số lớn nhất là ảnh hưởng nhiều nhất, TLT HAY TLN  phụ thuộc vào dấu + or -
#print(correlation)
import seaborn as sns   # thư viện hiển thị dữ liệu
sns.pairplot(data)
#sns.heatmap(correlation,annot=True)
```

```python

from sklearn import linear_model
model = linear_model.LinearRegression()       # hồi quy tuyến tính
model.fit(X=pd.DataFrame(data['YearsExperience']),y = data['Salary'])                     # dạy băng lệnh fit     X viết hoa , y viết thường
print(model.intercept_)
print(model.coef_)
model.score(X=pd.DataFrame(data['YearsExperience']),y = data['Salary']) # hàm score tính độ chính xác

data.plot(kind = 'scatter', x = 'YearsExperience', y='Salary',figsize = (9,9), color = 'black',xlim=(0,7))
y_pred=model.predict(X=pd.DataFrame(data['YearsExperience'])) 
plt.plot(data['YearsExperience'],y_pred,color = 'blue') 
print('độ chính xác của mô hình',model.score(X=pd.DataFrame(data['YearsExperience']),y = data['Salary']))
```

```python
model = linear_model.LinearRegression()
X=pd.DataFrame([data['YearsExperience'],data['YearsExperience']**2]).T
model.fit(X,y=data['Salary'])
print('intercept của mô hình',model.intercept_)
print('coef của mô hình',model.coef_)
print('độ chính xác của mô hình',model.score(X,y=data['Salary']))
data.plot(kind = 'scatter', x = 'YearsExperience', y='Salary',figsize = (9,9), color = 'black',xlim=(0,10))
range = np.arange(1.5,6.5,0.1)
predictor = pd.DataFrame([range,range**2]).T
y_pred=model.predict(X=predictor)
plt.plot(range,y_pred,color='blue')
```
